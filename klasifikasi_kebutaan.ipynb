```python
import pandas as pd
import csv
from google.colab import drive
drive.mount('/content/drive')
```

```python
# baca data di google drive
data_frame = pd.read_csv('/content/drive/My Drive/tubes_ml/dataset_kebutaan.csv')
```

```python
def prediksi(nama,umur,diabetes,hipertensi,intraokular):
  
  # total kelas ya dan tidak
  temp_target_ya = 0
  temp_target_tidak = 0
  jumlah_target = 33


  # status
  status = ""
  statusUmur = ""
  statusIntraokular = ""


  # data latih
  train_umur = data_frame['Umur']
  train_diabetes = data_frame['Diabetes']
  train_hipertensi = data_frame['Hipertensi']
  train_intraokular = data_frame['Intraokular'];
  train_target = data_frame['Kebutaan']


  # variabel penampung diabetes
  temp_diabetesYa = 0
  temp_diabetesTidak = 0

  # variabel penampung hipertensi
  temp_hipertensiYa = 0
  temp_hipertensiTidak = 0

  # variable penampung intraokular
  temp_rendah_ya = 0
  temp_rendah_tidak = 0
  temp_sedang_ya = 0
  temp_sedang_tidak = 0
  temp_tinggi_ya = 0
  temp_tinggi_tidak = 0
  intraokularHasilYa = 0
  intraokularHasilTidak = 0

  # variabel penampung umur
  temp_muda_ya = 0
  temp_muda_tidak = 0
  temp_parobaya_ya = 0
  temp_parobaya_tidak = 0
  temp_tua_ya = 0
  temp_tua_tidak = 0
  umurHasilYa = 0
  umurHasilTidak = 0


  # perhitungan probabilitas bersyarat / likelihood
  for i in range(0,32+1):

    # menghitung probabilitas prior untuk kelas Kebutaan YA dan Tidak
    if(train_target[i] == "YA"):
      temp_target_ya += 1
    else:
      temp_target_tidak += 1

    
    # diabetes
    if(diabetes == train_diabetes[i]):
      if(train_diabetes[i] == "YA"):
        if(train_diabetes[i] == train_target[i]):
            temp_diabetesYa += 1
        else:  
            temp_diabetesTidak += 1
      else:
        if(train_diabetes[i] == train_target[i]):
            temp_diabetesTidak += 1
        else:
            temp_diabetesYa += 1

    # hipertensi
    if(hipertensi == train_hipertensi[i]):
      if(train_hipertensi[i] == "YA"):
        if(train_hipertensi[i] == train_target[i]):
            temp_hipertensiYa += 1
        else:  
            temp_hipertensiTidak += 1
      else:
        if(train_hipertensi[i] == train_target[i]):
            temp_hipertensiTidak += 1
        else:
            temp_hipertensiYa += 1

  # intraokular
    
  if(intraokular < 21):
    for i in range(0,32+1):
      if(train_intraokular[i] < 21):
        if(train_target[i] == "YA"):
          temp_rendah_ya += 1
        else:
          temp_rendah_tidak += 1
  
    intraokularHasilYa = temp_rendah_ya
    intraokularHasilTidak = temp_rendah_tidak
    statusIntraokular = "RENDAH"

  elif(intraokular >= 21 and intraokular <= 41):
    for i in range(0,32+1):
      if(train_intraokular[i] >= 21 and train_intraokular[i] <= 41 ):
        if(train_target[i] == "YA"):
          temp_sedang_ya += 1
        else:
          temp_sedang_tidak += 1

    intraokularHasilYa = temp_sedang_ya
    intraokularHasilTidak = temp_sedang_tidak
    statusIntraokular = "SEDANG"    

  else:
    for i in range(0,32+1):
      if(train_intraokular[i] > 41):
        if(train_target[i] == "YA"):
          temp_tinggi_ya += 1
        else:
          temp_tinggi_tidak += 1

    intraokularHasilYa = temp_tinggi_ya
    intraokularHasilTidak = temp_tinggi_tidak
    statusIntraokular = "TINGGI"
    
  # umur
  if(umur >= 35 and umur <= 45):
    for i in range(0,32+1):
      if(train_umur[i] >= 35 and train_umur[i] <= 45):
        if(train_target[i] == "YA"):
          temp_muda_ya += 1
        else:
          temp_muda_tidak += 1
 
    umurHasilYa = temp_muda_ya
    umurHasilTidak = temp_muda_tidak
    statusUmur = "MUDA"
 
  elif(umur >= 45 and umur <= 55):
    for i in range(0,32+1):
      if(train_umur[i] >= 45 and train_umur[i] <= 55 ):
        if(train_target[i] == "YA"):
          temp_parobaya_ya += 1
        else:
          temp_parobaya_tidak += 1
   
    umurHasilYa = temp_parobaya_ya
    umurHasilTidak = temp_parobaya_tidak
    statusUmur = "PAROBAYA"
    
  elif(umur > 55):
    for i in range(0,32+1):
      if(train_umur[i] > 55):
        if(train_target[i] == "YA"):
          temp_tua_ya += 1
        else:
          temp_tua_tidak += 1
    
    umurTuaYa = temp_tua_ya
    umurTuaTidak = temp_tua_tidak
    umurHasilYa = temp_tua_ya
    umurHasilTidak = temp_tua_tidak
    statusUmur = "TUA"
    
  # perhitungan posterior kebutaan = YA
  prediksiButa = round((temp_target_ya/jumlah_target) * (umurHasilYa/temp_target_ya) * (temp_diabetesYa/temp_target_ya) * (temp_hipertensiYa / temp_target_ya) * (intraokularHasilYa/temp_target_ya), 3)
  # perhitungan posterior kebutaan = TIDAK
  prediksiTidakButa = round((temp_target_tidak/jumlah_target) * (umurHasilTidak/temp_target_tidak) * (temp_diabetesTidak/temp_target_tidak) * (temp_hipertensiTidak/temp_target_tidak) * (intraokularHasilTidak/temp_target_tidak), 3)

  # Bandingkan nilai prediksi Buta dan Tidak buta (nilai maksimum menentukan prediksi)
  if(prediksiButa > prediksiTidakButa):
    status = "BUTA"
  else:
    status = "TIDAK BUTA"

  print(f"""
  Nama Pasien        : {nama}
  Umur               : {umur}
  Diabetes           : {diabetes}
  Hipertensi         : {hipertensi}
  Intraokular        : {intraokular}
  ----------------------------------
  UMUR (ğ‘¥1)
  ğ‘ƒ(ğ‘¥1|ğ¶) = ğ‘ƒ({statusUmur}|KEBUTAAN â€œYAâ€) = {umurHasilYa}/{temp_target_ya} = {round(umurHasilYa/temp_target_ya, 3)}
  ğ‘ƒ(ğ‘¥1|ğ¶) = ğ‘ƒ({statusUmur}|KEBUTAAN â€œTIDAKâ€) = {umurHasilTidak}/{temp_target_tidak} = {round(umurHasilTidak/temp_target_tidak, 3)}

  Diabetes (ğ‘¥2)
  ğ‘ƒ(ğ‘¥2|ğ¶) = ğ‘ƒ({diabetes}|KEBUTAAN â€œYAâ€) = {temp_diabetesYa}/{temp_target_ya} = {round(temp_diabetesYa/temp_target_ya, 3)}
  ğ‘ƒ(ğ‘¥2|ğ¶) = ğ‘ƒ({diabetes}|KEBUTAAN â€œTIDAKâ€) = {temp_diabetesTidak}/{temp_target_tidak} = {round(temp_diabetesTidak/temp_target_tidak, 3)}

  Hipertensi (ğ‘¥3)
  ğ‘ƒ(ğ‘¥3|ğ¶) = ğ‘ƒ({hipertensi}|KEBUTAAN â€œYAâ€) = {temp_hipertensiYa}/{temp_target_ya} = {round(temp_hipertensiYa / temp_target_ya, 3)}
  ğ‘ƒ(ğ‘¥3|ğ¶) = ğ‘ƒ({hipertensi}|KEBUTAAN â€œTIDAKâ€) = {temp_hipertensiTidak}/{temp_target_tidak} = {round(temp_hipertensiTidak / temp_target_tidak, 3)}

  INTRAOKULAR (ğ‘¥4)
  ğ‘ƒ(ğ‘¥4|ğ¶) = ğ‘ƒ({statusIntraokular}|KEBUTAAN â€œYAâ€) = {intraokularHasilYa}/{temp_target_ya} = {round(intraokularHasilYa/temp_target_ya, 3)}
  ğ‘ƒ(ğ‘¥4|ğ¶) = ğ‘ƒ({statusIntraokular}|KEBUTAAN â€œTIDAKâ€) = {intraokularHasilTidak}/{temp_target_tidak} = {round(intraokularHasilTidak/temp_target_tidak, 3)}

  KEBUTAAN
  N = 33
  ğ‘ƒ(ğ¶) = Nj/N = ğ‘ƒ(YA) = {temp_target_ya}/{jumlah_target} = {round(temp_target_ya/jumlah_target, 3)} 
  ğ‘ƒ(ğ¶) = Nj/N = ğ‘ƒ(TIDAK) = {temp_target_tidak}/{jumlah_target} = {round(temp_target_tidak/jumlah_target, 3)}
  
  ğ‘ƒ(ğ¶|ğ‘¥) = ğ‘ƒ(ğ‘¥|ğ¶) * ğ‘ƒ(ğ¶)/ğ‘ƒ(ğ‘¥)
  Kebutaan = YA
  ğ‘ƒ(YA|ğ‘¥)  = {round(temp_target_ya/jumlah_target, 3)} * {round(umurHasilYa/temp_target_ya, 3)} * {round(temp_diabetesYa/temp_target_ya, 3)} * {round(temp_hipertensiYa/temp_target_ya, 3)} * {round(intraokularHasilYa/temp_target_ya, 3)} 
           = {prediksiButa}

  Kebutaan   = TIDAK
  ğ‘ƒ(TIDAK|ğ‘¥) = {round(temp_target_tidak/jumlah_target, 3)} * {round(umurHasilTidak/temp_target_tidak, 3)} * {round(temp_diabetesTidak/temp_target_tidak, 3)} * {round(temp_hipertensiTidak/temp_target_tidak, 3)} * {round(intraokularHasilTidak/temp_target_tidak, 3)}
             = {prediksiTidakButa}
        
  Maka, Pasien yang bernama {nama} Hasil Prediksinya : {status}
  
  """)
```
